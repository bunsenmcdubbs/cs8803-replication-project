```python
from glob import glob
train_data_path = 'data/data-from-authors/train_data/*.json'
train_data_filenames = glob(train_data_path)
```

```python
import sys
import os
from importlib import reload
import preproc
from src.preproc import wikidata
```

```python
reload(preproc);
reload(wikidata);
```

```python
processed_train_data_dir = 'data/processed/train_data'
cache_hits = [wikidata.search_cache_hits]
cache_misses = [wikidata.search_cache_misses]

print('Total: {} files'.format(len(train_data_filenames)))
for i, filepath in enumerate(train_data_filenames):
    infilename = os.path.basename(filepath)
    outfilename = os.path.join(processed_train_data_dir, infilename)
    if os.path.isfile(outfilename):
        continue

    print('Processing #{}: {}'.format(i + 1, infilename))

    infile = open(filepath, 'r')
    outfile = open(outfilename, 'w+')
    
    preproc.preprocess(infile, outfile, is_parsed=True)
    cache_hits.append(wikidata.search_cache_hits)
    cache_misses.append(wikidata.search_cache_misses)
    hits = cache_hits[-1] - cache_hits[-2]
    misses = cache_misses[-1] - cache_misses[-2]
    print('cache hit rate: {:.2f}%'.format(100.*hits/(hits + misses)),
          '\tcache hits:', hits,
          '\tcache misses:', misses)
    print()
    
    infile.close()
    outfile.close()
```
