```python
mpqa_data_path = 'data/database.mpqa.3.0/'
with open('data/mpqa_used_files_filenames.txt', 'r') as f:
    mpqa_data_filepaths = [
        os.path.join(mpqa_data_path, filepath.strip())
        for filepath in f
    ]
processed_mpqa_data_dir = 'data/processed/mpqa'
```

```python
filepaths = mpqa_data_filepaths
out_dir = processed_mpqa_data_dir
```

```python
cache_hits = [wikidata.search_cache_hits]
cache_misses = [wikidata.search_cache_misses]

print('Total: {} files'.format(len(filepaths)))
for i, filepath in enumerate(filepaths):
    infilename = os.path.basename(filepath)
    outfilename = os.path.join(out_dir, infilename)
    if os.path.isfile(outfilename):
        continue

    print('Processing #{}: {}'.format(i + 1, infilename))

    infile = open(filepath, 'r')
    outfile = open(outfilename, 'w+')
    
    preproc.preprocess(infile, outfile)
    cache_hits.append(wikidata.search_cache_hits)
    cache_misses.append(wikidata.search_cache_misses)
    hits = cache_hits[-1] - cache_hits[-2]
    misses = cache_misses[-1] - cache_misses[-2]
    print('cache hit rate: {:.2f}%'.format(100.*hits/(hits + misses)),
          '\tcache hits:', hits,
          '\tcache misses:', misses)
    print()
    
    infile.close()
    outfile.close()
```
